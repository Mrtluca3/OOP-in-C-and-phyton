```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
data = 'sim3.dat'
params = 'params3.ini'

datContent = [[float(i.strip().split()[j]) for j in range(len(i.strip().split()))] for i in open(data).readlines()]
datContent = np.array(datContent)

paramsContent = [[float(i.strip().split()[j]) for j in range(len(i.strip().split()))] for i in open(params).readlines()]


#Plot the different graphs for different simulations
#Graph the satellite's trajectory
if paramsContent[0][-1] == 0:
    model = "RungeKutta2"
else:
    model = "Euler"
dt = str(paramsContent[0][-3])

fig = plt.figure(figsize=(10,5))
fig.suptitle("Trajectory of satellite computed with " + model + " and steps dt=" + dt)

ax = fig.add_subplot(1,2,1)
ax.plot(datContent[:,1], datContent[:,0], '-', label='satellite')
ax.set_xlabel("x [m]")
ax.set_ylabel("y [m]")

R0=6371e3
ax.plot(R0*np.cos(np.linspace(0,2*3.1415, 3000)),R0*np.sin(np.linspace(0,2*3.1415, 3000)), label='Earth')
ax.legend()

#Plot the satellite energy in time
ax2 = fig.add_subplot(1,2,2)
m=1200
M=5.972e24
G=6.67e10-11

E = np.linalg.norm(np.array(datContent[:,3:])*float(dt), axis=1)**2*m/2 - G*m*M/np.linalg.norm(datContent[:,:3], axis=1)
ax2.plot(np.arange(0, len(E)), E)
ax2.set_xlabel("t [s]")
ax2.set_ylabel("Energy [J]")
fig.tight_layout()

print('Does z gets different from 0? z_max: ', max(datContent[:,3]))
```

```python
data = 'sim.dat'
params = 'params.ini'

datContent = [[float(i.strip().split()[j]) for j in range(len(i.strip().split()))] for i in open(data).readlines()]
datContent = np.array(datContent)

paramsContent = [[float(i.strip().split()[j]) for j in range(len(i.strip().split()))] for i in open(params).readlines()]


#Plot the different graphs for different simulations
#Graph the satellite's trajectory
if paramsContent[0][-1] == 0:
    model = "RungeKutta2"
else:
    model = "Euler"
dt = str(paramsContent[0][-3])

fig = plt.figure(figsize=(10,5))
fig.suptitle("Free fall of satellite computed with " + model + " and steps dt=" + dt)

ax = fig.add_subplot(1,2,1)

g=9.81
t=np.arange(0, len(datContent[datContent[:,0] > 6.371e6, 0]))*float(dt)

y=-g*t**2/2+(paramsContent[0][0])
ax.plot(y, datContent[datContent[:,0] > 6.371e6, 0] - 6.371e6, '-', label='satellite')
ax.set_xlabel("Theoric x [m]")
ax.set_ylabel("Computed x [m]")

ax.legend()

#Plot the satellite energy in time
ax2 = fig.add_subplot(1,2,2)
m=1200
M=5.972e24
G=6.67e10-11

E = np.linalg.norm(np.array(datContent[:,3:])*float(dt), axis=1)**2*m/2 - G*m*M/np.linalg.norm(datContent[:,:3], axis=1)
ax2.plot(np.arange(0, len(E)), E)
ax2.set_xlabel("t [s]")
ax2.set_ylabel("Energy [J]")
fig.tight_layout()

print('Does z gets different from 0? z_max: ', max(datContent[:,3]))
```

The confrontation between theoric and computed x(t) is a straight line, as should be. The energy tends to go a bit down, probably due to computational errors.
